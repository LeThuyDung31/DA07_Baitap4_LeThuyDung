```python
import csv
csv_path = "C:\\Users\\Admin\\Desktop\\btnop\\supermarket_sales_vn.csv"
city = []
gender = []
price = []
with open(csv_path, mode='r', encoding="utf-8", newline='') as f:
    
    f_csv = csv.reader(f)
    header = next(f_csv)
    for i, row in enumerate(f_csv):
        city.append(row[1])
        gender.append(row[3])
        price.append(row[5])
```

```python
count_HN = 0
count_TPHCM = 0
count_DANANG = 0 
for tp in city:
    if tp == "Hà Nội": 
        count_HN += 1
    elif tp == "TP HCM":
        count_TPHCM += 1
    elif tp == "Đà Nẵng":
        count_DANANG += 1
print("Hà Nội:",count_HN)
print("TP HCM:",count_TPHCM)
print("Đà Nẵng:",count_DANANG)    
```

```python
count_nam = 0
count_nu = 0
for i in gender :
    if i == "Nam":
        count_nam += 1
    else:
        count_nu += 1
print ("Nam:",count_nam)
print ("Nữ:", count_nu)
```

```python

import matplotlib
import matplotlib.pyplot as plt
#print(matplotlib)
from matplotlib import pyplot as plt
```

### Tần số

```python
#tần số thành phố
ds_city = [count_HN,count_TPHCM,count_DANANG]
ds_gender = [count_nam,count_nu]

label_city = ["Hà Nội","Hồ Chí Minh","Đà Nẵng"]
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(2,2,1)
ax1.bar(x=range(len(ds_city)),height= ds_city,width=0.4)
ax1.set_xticks(range(len(label_city)),label_city)
ax1.set_title("Biểu đồ thể hiện tần số địa phương",size=10,weight="bold")
for index in range(len(ds_city)):
    ax1.text((index-0.15),(ds_city[index]-50),ds_city[index],size=14,color="w",fontweight="bold")

#tần số gender
label_gender = ["Nam","Nữ"]    
ax2 = fig.add_subplot(2,2,2)  
ax2.bar(x=range(len(ds_gender)),height=ds_gender,width=0.3,color="pink")
ax2.set_xticks(range(len(label_gender)),label_gender) 
ax2.set_title("Biểu đồ thể hiện tần số giới tính",size=10,weight="bold")
for index in range(len(ds_gender)):
    ax2.text((index-0.1),(ds_gender[index]-55),ds_gender[index],size=14,color="brown",fontweight="bold")

plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(2,2,1)
labels = label_city
ax1.pie((ds_city), autopct="%.1f%%",
        shadow= True,
        explode=[0.1,0,0],
        colors=["cyan","pink","lightgreen"])
ax1.set_title("Biểu đồ tròn thể hiện tần số địa phương",size=10,weight="bold",color="orchid")
plt.legend(labels, loc=[0.78,0.7])

ax2 = fig.add_subplot(2,2,2)
ax2.pie(ds_gender,labels=label_gender, autopct='%1.1f%%',
        colors=["cyan","pink"],
       
        shadow=True, 
        startangle=90)
ax2.set_title("Biểu đồ tròn thể hiện tần số giới tính",size=10,weight="bold",color="cyan",)
plt.legend(label_gender, loc=[0.78,0.7])
plt.show()
```

## Tần suất

```python
tan_suat_city = []
tan_suat_gender = []
for i in set(city):
    count = 0
    for x in city:
        if x == i: count += 1
    ty_le = count*100/len(city)
    tan_suat_city.append(ty_le)
for i in set(gender):
    count = 0
    for y in gender:
        if y == i: count += 1
    ty_le = count*100/len(gender)
    tan_suat_gender.append(ty_le)
tan_suat_city
tan_suat_gender
```

```python
#tần suất thành phố
fig = plt.figure(figsize=(10,8)) 
ax3 = fig.add_subplot(2,2,3)
label_city_tsuat= ["Hà Nội","Đà Nẵng","Hồ Chí Minh"]
ax3.bar(x=range(len(tan_suat_city)),height=tan_suat_city,color="Magenta",width=0.5,edgecolor = "k")
ax3.set_xticks(range(3),label_city_tsuat)
ax3.set_title("Biểu đồ cột thể hiện tần suất địa phương",size=10,fontweight="bold",color="black",)
for index in range(len(tan_suat_city)):
    ax3.text((index-0.2),(tan_suat_city[index]-5),tan_suat_city[index],size=14,color="w",fontweight="bold")

#tần suất giới tính    
ax4 = fig.add_subplot(2,2,4)
ax4.bar(x=range(len(tan_suat_gender)),height=tan_suat_gender,color="Magenta",width=0.5,edgecolor = "k")
ax4.set_xticks(range(2),label_gender)
ax4.set_title("Biểu đồ cột thể hiện tần suất giới tính",size=10,fontweight="bold",color="black",)
for index in range(len(tan_suat_gender)):
    ax4.text((index-0.1),(tan_suat_gender[index]-5),tan_suat_gender[index],size=14,color="w",fontweight="bold")
plt.show()
```
