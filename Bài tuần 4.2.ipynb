```python
import csv
csv_path = "C:\\Users\\Admin\\Desktop\\btnop\\supermarket_sales_vn.csv"
city = []
gender = []
price = []
with open(csv_path, mode='r', encoding="utf-8", newline='') as f:
    
    f_csv = csv.reader(f)
    header = next(f_csv)
    for i, row in enumerate(f_csv):
        city.append(row[1])
        gender.append(row[3])
        price.append(row[5])
```

```python
count_HN = 0
count_TPHCM = 0
count_DANANG = 0 
for tp in city:
    if tp == "Hà Nội": 
        count_HN += 1
    elif tp == "TP HCM":
        count_TPHCM += 1
    elif tp == "Đà Nẵng":
        count_DANANG += 1
print("Hà Nội:",count_HN)
print("TP HCM:",count_TPHCM)
print("Đà Nẵng:",count_DANANG)    
```

```python
count_nam = 0
count_nu = 0
for i in gender :
    if i == "Nam":
        count_nam += 1
    else:
        count_nu += 1
print ("Nam:",count_nam)
print ("Nữ:", count_nu)
```

```python
import matplotlib
from matplotlib import pyplot as plt
print(matplotlib.__version__)
```

### Tần số

```python
ds_city = [count_HN,count_TPHCM,count_DANANG]
ds_gender = [count_nam,count_nu]

ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)

ax1.bar(x=list(set(city)),height= ds_city,width=0.4)
ax2.bar(x=list(set(gender)),height=ds_gender,width=0.3,color="pink")

plt.show()
```

## Tần suất

```python
tan_suat_city = []
for i in set(city):
    count = 0
    for x in city:
        if x == i: count += 1
    ty_le = count*100/len(city)
    tan_suat_city.append(ty_le)

tan_suat_city   
```

```python
tan_suat_gender = []
for i in set(gender):
    count = 0
    for y in gender:
        if y == i: count += 1
    ty_le = count*100/len(gender)
    tan_suat_gender.append(ty_le)
    print(ty_le)
tan_suat_gender   
```

```python

ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)

ax1.bar(x= list(set(city)),height= tan_suat_city,width=0.4)
ax2.bar(x=list(set(gender)),height=tan_suat_gender,width=0.3,color="pink")

plt.show()
```
